```python
import numpy as np
```

```python
from scipy.integrate import odeint
```

```python
import matplotlib.pyplot as plt
```

# Exponential growth is continuous population growth in an environment where resources are unlimited, it is density independent growth

```python
# Change in population size
def growth(P, time):
    tau = 2 # proportionality constant or time constant
    dPdt =  P/tau 
    return dPdt
```

```python
# Original population size
P_zero = 540000000
```

```python
# time points 
time = np.linspace(0,20)
```

```python
# solve ODE
Population = odeint(growth, P_zero, time)
```

```python
# plot results
plt.plot(time, Population)
plt.xlabel('time')
plt.ylabel('Population')
plt.show()
```

# Logistic growth is continuous population growth in an environment where resources are limited; it is density dependent growth 

```python
# Change in population size
def growth(P, time):
    tau = 2 # proportionality constant or time constant
    K = 20000 # maximum allowble size of the population 
    dPdt =  (P * (1 - P/K))/tau
    return dPdt
```

```python
# Original population size
P_zero = 200
```

```python
# time points 
time = np.linspace(0,20)
```

```python
# solve ODE
Population = odeint(growth, P_zero, time)
```

```python
# plot results
plt.plot(time, Population)
plt.xlabel('time')
plt.ylabel('Population')
plt.show()
```
